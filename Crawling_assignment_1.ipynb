```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=103&oid=032&aid=0002930147')
sun = BeautifulSoup(html, "html.parser")
```

```python
sun
```

articleBody를 가져 오게 되면 댓글과 광고의 text까지 전부 가져 오게 되어 기사만을 가져 오기 위하여 articleBodyContents를 가져왔습니다.

```python
div = sun.find("div",{"id":"articleBodyContents"})
article = div.get_text()
print(article)
```

기사 본문 위에 필요없는 부분이 있어 스플릿 함수로 뺐습니다.

```python
article = article.strip().split("\n")[-1]
```

```python
article 
```

기사 본문의 있는 단어들을 모두 가져와 갯수를 세도록 하였습니다.

```python
all_words = article.split(" ")

word_dict = {}
for word in all_words:
    if word in word_dict.keys():
        word_dict[word]+=1
    else:
        word_dict[word] = 1
```

```python
# Create a file to write to, add headers row
with open('Crawling_assignment_1.csv', 'w',encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(['word', 'countNum'])
    for key, val in word_dict.items():
        writer.writerow([key,val])
        
```
